```python
# download data from https://www.kaggle.com/paramaggarwal/fashion-product-images-small
import pandas as pd
```

```python
df = pd.read_csv("styles.csv", nrows=10000, error_bad_lines=False)
image_df = pd.read_csv("images.csv")
df['filename'] = df.apply(lambda row: str(row['id']) + ".jpg", axis=1)

# merge images.csv and styles.csv together
df = pd.merge(df, image_df, on = ["filename"])
df = df.sample(frac=1).reset_index(drop=True)

df.head(10)
```

```python
df_top = df[df['subCategory'] == 'Topwear']
df_bottom = df[df['subCategory'] == 'Bottomwear']
df_shoe = df[df['subCategory'] == 'Shoes']
df_belt = df[df['subCategory'] == 'Belts']
df_sock = df[df['subCategory'] == 'Socks']
df_dress = df[df['subCategory'] == 'Dress']
```
